### 5.3 APLICACION DE MODELOS DE ML

#### **Aplicamos el metodo Arbol de decision para nuestro primer objetivo**

---


Aplicamos nuestro primero modelo de prediccion.

---
Arbol de desicion (deph=6)



```python
# Codificar las etiquetas como valores binarios
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
Y_encoded = label_encoder.fit_transform(Y)
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba

X_train, X_test, y_train, y_test = train_test_split(X_NEW, Y_encoded, test_size=0.33, random_state=42)

# Entrenar el arbol

clf = tree.DecisionTreeClassifier(max_depth=6, min_samples_leaf=5, random_state=4)
clf = clf.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred= clf.predict(X_test)

```

Metricas

---



```python
# Calcular métricas de evaluación

classification_rep = classification_report(y_test, y_pred)
test_accuracy = accuracy_score(y_test, y_pred)#Calculo el accuracy en Test
conf_matrix = confusion_matrix(y_test,y_pred) #matriz de confusion


# Visualizar las métricas
print(f"precisión: {test_accuracy}")
print("Matriz de confusión:")
print(conf_matrix)
print("Reporte de clasificacion:")
print(classification_rep)
```

matriz de confusion

```python
# Visualizar la matriz de confusión para el modelo de Árboles de Decisión
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap="Blues", fmt="d")
plt.title("Confusion Matrix - Decision Tree Model")
plt.show()
```

Interpretacion


---


El modelo tiene un rendimiento general muy bueno, con una precisión global del 99%. Esto sugiere que el modelo es capaz de clasificar correctamente la gran mayoría de las instancias.

```python

# Graficar el árbol de decisión
fig = plt.figure(figsize=(18,10))
_ = tree.plot_tree(clf, feature_names=X_train.columns,
                   class_names=label_encoder.classes_,
                   filled=True)
```

```python
# Obtener las probabilidades de predicción para la clase positiva
y_pred_proba = clf.predict_proba(X_test)[:, 1]

from sklearn.metrics import roc_curve, auc, accuracy_score, confusion_matrix, classification_report
# Calcular la curva ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Graficar la curva ROC
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
```
